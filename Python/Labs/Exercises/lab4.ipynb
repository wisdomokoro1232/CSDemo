```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns


google_data = pd.read_csv("C:/Users/wwo20/OneDrive - Imperial College London/Documents/TAP/CSDemo/Python/Labs/Data/GOOGL.csv")
google_data.head()
```

```python
from matplotlib.dates import YearLocator, DateFormatter

fig = plt.figure(figsize=(12,8))
plt.rcParams['xtick.major.size'] = 0.5
google_data['Date'] = pd.to_datetime(google_data['Date'])

# Main plot - takes up the entire figure
ax1 = fig.add_subplot(111)
ax1.plot(google_data['Date'], google_data['Close'], label='GOOGL Close Price', color='red')
ax1.set_xlabel('Date')
ax1.set_ylabel('Close Price ($)')
ax1.set_title('Google Stock Price')
ax1.xaxis.set_major_locator(YearLocator(1))
ax1.xaxis.set_major_formatter(DateFormatter("%Y"))

# Overlay plot - positioned in upper left corner
# add_axes([left, bottom, width, height]) - all values are fractions of figure size
ax2 = fig.add_axes([0.15, 0.55, 0.5, 0.25])  # [x, y, width, height] in figure coordinates
ax2.plot(google_data['Date'], google_data['Volume'], color='blue', linewidth=0.6)
ax2.set_title('Volume', fontsize=8)
ax2.tick_params(axis='both', which='major', labelsize=6)
ax2.xaxis.set_major_locator(YearLocator(1))  # Show every 2 years for less crowding
ax2.xaxis.set_major_formatter(DateFormatter("%Y"))
#remove gridlines
ax2.grid(False)
ax1.grid(False)
plt.show()
```

```python
from matplotlib.dates import YearLocator, DateFormatter,MonthLocator
import datetime 

fig = plt.figure(figsize=(14,8))
plt.rcParams['xtick.major.size'] = 0.5
google_data['Date'] = pd.to_datetime(google_data['Date'])

# Main plot - takes up the entire figure
ax1 = fig.add_subplot(111)
ax1.plot(google_data['Date'], google_data['Close'], label='GOOGL Close Price', color='red')
ax1.set_xlabel('Date')
ax1.set_ylabel('Close Price ($)')
ax1.set_title('Google Stock Price')
ax1.xaxis.set_major_locator(YearLocator(1))
ax1.xaxis.set_major_formatter(DateFormatter("%Y"))

# Overlay plot - positioned in upper left corner
# add_axes([left, bottom, width, height]) - all values are fractions of figure size
ax2 = fig.add_axes([0.58, 0.15, 0.33, 0.18])  # [x, y, width, height] in figure coordinates
ax2.plot(google_data[google_data['Date'].dt.strftime('%Y') == '2017' ]['Date'], google_data[google_data['Date'].dt.strftime('%Y') == '2017']['Volume'], color='blue', linewidth=0.6)
ax2.set_title('Volume', fontsize=8)
# ax2.tick_params(axis='both', which='major', labelsize=6)
# ax2.xaxis.set_major_locator(MonthLocator(1))  # Show every 2 years for less crowding
#Shpow month name not number
ax2.xaxis.set_major_formatter(DateFormatter("%b"))
#remove gridlines
ax2.grid(False)
ax1.grid(False)
plt.show()
```
